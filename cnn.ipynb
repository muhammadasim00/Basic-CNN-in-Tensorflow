```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras import layers, models

# Define the CNN architecture
model = models.Sequential([
    # Convolutional layers
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(30, 30, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    # Flatten the feature maps
    layers.Flatten(),
    # Fully connected layers
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')  # Output layer with softmax activation for 10 classes
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Print the model summary
model.summary()

# Load and preprocess the selected image
img_path = 'ahsan.png'  # Replace 'path_to_your_selected_image.jpg' with the path to your selected image file
img = image.load_img(img_path, target_size=(30, 30), color_mode='grayscale')  # Resize image to match input shape (28x28 grayscale)
img_array = image.img_to_array(img)
img_array /= 255.0  # Normalize pixel values to [0, 1]
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

# Display the input image
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Input Image')
plt.show()

# Predict the class probabilities
predictions = model.predict(img_array)

# Get the predicted class label
predicted_class = np.argmax(predictions[0])

# Print the predicted class label
print(f'Predicted class: {predicted_class}')
```
